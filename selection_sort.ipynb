```python
def binarySearch(ilist, item):
    first = 0
    last = len(ilist)-1
    found = False

    while first<=last and not found:
        midpoint = (first + last)//2
        print(ilist[midpoint] + " - ", "index: " + str(midpoint))
        if ilist[midpoint] == item:
            found = True
        else:
            if item < ilist[midpoint]:
                last = midpoint-1
            else:
                first = midpoint+1

    return found

s_list = ["babka","baklava","cheesecake","cupcake","danish","eclair","funnelcake","kringle","lamington","profiterole","sopaipilla","strudel","tiramisu","torte","turnover"]
print(binarySearch(s_list, "doughnut"))

print(binarySearch(s_list, "tiramisu"))
```

```python
def bubbleSort(ilist):
    for num in range(len(ilist)-1,0,-1):
        for i in range(num):
            if ilist[i]>ilist[i+1]:
                temp = ilist[i]
                print("temp becomes item at ilist[i]=>> " + str(temp))
                
                ilist[i] = ilist[i+1]
                print("item at ilist[i] becomes ilist[i+1] =>> " + str(ilist[i+1]))
                
                ilist[i+1] = temp
                
                print("Then ilist[i+1] becomes temp =>> " + str(temp))
                
                print(ilist, "\n\n")


ilist = [9,20,6,10,14,8,60,11]

print("show the series of steps taken by the Bubble Sort algorithm while sorting this list.\n\n\n")
bubbleSort(ilist)
print("sorted list = ", ilist)
```

```python
def selectionSort(ilist):
   for fslot in range(len(ilist)-1,0,-1):
       positionOfMax=0
       for location in range(1,fslot+1):
           if ilist[location]>ilist[positionOfMax]:
               positionOfMax = location

       temp = ilist[fslot]
       ilist[fslot] = ilist[positionOfMax]
       ilist[positionOfMax] = temp
       print(ilist, "\n")

alist = [9,20,6,10,14,8,60,11]
print("show the series of steps taken by the Selection Sort algorithm while sorting this list.\n\n")
selectionSort(alist)
print(ilist)
```
